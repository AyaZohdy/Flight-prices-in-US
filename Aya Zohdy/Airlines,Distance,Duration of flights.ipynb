# Airlines & Prices correlation

**The column**

```python
#df['segmentsAirlineName'] = df['segmentsAirlineName']#.astype('category')
df['segmentsAirlineName']
```

**1. cleaning the data**

```python
#Standardizing Airline Names

airline_mapping = {
    'American Airlines||American Airlines': 'American Airlines',
    'Delta||Delta': 'Delta',
    'United||United': 'United',
    'Alaska Airlines||Alaska Airlines': 'Alaska Airlines'
}

# Replace variations with standard names
df['segmentsAirlineName'] = df['segmentsAirlineName'].replace(airline_mapping)
```

**2. Count the values**

```python
counts = df['segmentsAirlineName'].value_counts()
counts
```

**3. Filtring the data by making knew columns for the new values 'Top 3 airlines'**

```python
fillter = (df['segmentsAirlineName']=='American Airlines') | (df['segmentsAirlineName']=='Delta')|(df['segmentsAirlineName']=='United') 
top3_airlines = df[fillter]

top3_airlines['segmentsAirlineName'].value_counts()
```

**4. Change the type of the values**

```python
top3_airlines['segmentsAirlineName'] = top3_airlines['segmentsAirlineName'].astype('category')
top3_airlines['segmentsAirlineName']
```

**5. Show the unique values**

```python
top3_airlines.segmentsAirlineName.unique()
```

**6. Inplace the new dataframe**

```python
top3_airlines['totalFare'].dropna(inplace = True)
```
