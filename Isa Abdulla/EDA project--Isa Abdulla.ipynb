**12. counts how many flight segments**

```python
df['segment_count'] = df['segmentsDepartureAirportCode'].str.count(r'\|\|') + 1
df['stop_count'] = df['segment_count'] - 1
```

```python
def label_stops(x):
    if x == 0:
        return 'Nonstop'
    elif x == 1:
        return '1 Stop'
    else:
        return '2+ Stops'

df['stop_type'] = df['stop_count'].apply(label_stops)
```
